#Visualizing sales based on holidays

```python
import numpy as np
import sklearn 
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

holidays = pd.read_csv(
    "data/holidays_events.csv", usecols = ['date', 'type', 'locale'])

transactions =   pd.read_csv(
    "data/transactions.csv",  usecols=['date', 'store_nbr', 'transactions'])
```

```python
#Lets visualize the data
holidays.iloc[0:100, 0]
```

```python
transactions.iloc[0:10, 2]
```

```python
#Merge datasets on dates
merged_df = pd.merge(transactions, holidays, on='date', how='left')

avg_transactions = merged_df.groupby('type')['transactions'].mean()

# Plot average transactions by day type
avg_transactions.plot(kind='bar')
plt.title('Average Transactions by Day Type')
plt.xlabel('Day Type')
plt.ylabel('Average Transactions')
plt.show()
```

```python
#Let's plot transactions by the day of the month
transactions['date'] = pd.to_datetime(transactions['date'])
# transactions[0:10]

transactions['day_of_month'] = transactions['date'].dt.day
avg_transactions = transactions.groupby('day_of_month')['transactions'].mean().reset_index()

plt.plot(avg_transactions['day_of_month'], avg_transactions['transactions'], marker='o')
plt.title('Average Number of Transactions by Day of Month')
plt.xlabel('Day of Month')
plt.ylabel('Average Number of Transactions')
plt.xticks(range(1, 32))
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#Attempting other models

```python
oil_prices = pd.read_csv("data/oil.csv", usecols = ['date', 'dcoilwtico'])
```

```python
oil_prices.iloc[0:100, 1]
```

```python
oil_prices['date'] = pd.to_datetime(oil_prices['date'])

avg_transactions = transactions.groupby('date')['transactions'].mean().reset_index()

merged_df_oil = pd.merge(avg_transactions, oil_prices, on='date', how='left')

merged_df_oil = merged_df_oil.dropna(subset=['dcoilwtico'])

```

```python
merged_df_oil.iloc[0:100, 2]
```

```python
merged_df_oil.plot(x='dcoilwtico', y='transactions', kind='bar')
plt.title('Average Number of Transactions vs Oil Price')
plt.xlabel('Oil Price')
plt.ylabel('Average Number of Transactions')
plt.show()
```
